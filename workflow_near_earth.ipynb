```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/Near-Earth_Comets_-_Orbital_Elements.csv")
data_header = data.columns
```

```python
dc = list(data.columns)

count = data.count()

values = []
for c in count:
    values.append(c)

y_pos = np.arange(len(dc))

plt.bar(y_pos, values)
plt.xticks(y_pos, dc, rotation=90)

plt.show() 
```

```python
perc_m = (100*((max(values)*len(dc)) - sum(values)))/(max(values)*len(dc))
perc_e = 100-perc_m


labels = 'Missing', 'Present'
sizes = [perc_m, perc_e]
explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
data = data.dropna(axis = 1, how='all')

data1 = data.iloc[:, 1:15]

data1 = data1.interpolate(method ='linear', limit_direction ='both', axis=1)

result = pd.concat([data.iloc[:,:1], data1, data.iloc[:,15:]], axis=1)
```

```python
dc = list(result.columns)

count = result.count()

values = []
for c in count:
    values.append(c)

y_pos = np.arange(len(dc))

plt.bar(y_pos, values)
plt.xticks(y_pos, dc, rotation=90)

plt.show() 
```
